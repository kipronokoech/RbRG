```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import pickle
```

```python
#RECONSTRUCTION OF GROUND TRUTH LABELS
```

```python
im2=cv.imread("1-10-0010.jpg")
im2=cv.cvtColor(im2,cv.COLOR_BGR2RGB)
#pt = np.array(pickle.load(open("1-10-0010.npy","rb")))
pts=np.array(np.load("1-10-0010.npy",allow_pickle=True))
#pts=pt[::, ::-1] 
pt_l=pts.tolist()
for i in range (len(pts)):
    for j in range (65):
        pt_l[i][j].reverse()
ptt=np.array(pt_l)
img = cv.polylines(im2,np.int32(ptt),True,(255,0,0),8)
imgg=cv.cvtColor(img,cv.COLOR_BGR2RGB)
cv.imwrite("1-10-0010_truth.jpg",imgg)
plt.figure(figsize=(15,15))
plt.imshow(img)
```
